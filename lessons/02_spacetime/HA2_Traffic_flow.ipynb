```python
import numpy                       
from matplotlib import pyplot    
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16

T = 6 # min, final time

nx = 51 # number of grid points along x axis
L = 11. # km, a stretch of road 11 kilometers long 
rho_max = 250 # cars/km maximum traffic density
v_max = 80. # km/h, speed limit on a given distance l
dt = .001 # hours, time step 
dx = L/(nx-1)

nt = int(T*60./3600./dt)  #number of steps

# Use the following initial conditions for traffic density:
x = numpy.linspace(0,L,nx)
rho0 = numpy.ones(nx)*10
rho0[10:20] = 50

# IVP: rho(0,t) = 10

# plot initial density (t=0)
pyplot.figure(figsize=(8,4))  
pyplot.plot(x, rho0, color='#003366', ls='--', lw=3)
pyplot.ylim(0,60);

# create  rho at t=0
rho = rho0.copy()

# create an array with v_max, needed for the nonlinear coefficient
v_max_vect = numpy.ones(nx)*v_max

for n in range(1, nt):  
    rhon = rho.copy()
    un = v_max_vect - 2*rho/rho_max
    rho[1:] = rhon[1:]-un[1:]*dt/dx*(rhon[1:]-rhon[0:-1]) 
    rho[0] = 10.0
    
# plot the result density (t=0)
pyplot.figure(figsize=(8,4))  
pyplot.plot(x, rho, color='#003366', ls='--', lw=3)
pyplot.ylim(0,60);   


# some function for the output
# minimum velocity

def Vmin(r):
    """
    calculate minimum velocity for a given density r
    r: array
    returns: velocity in m/s
    """
    return v_max*(1-max(r)/rho_max)/3.6
# average density
def rhoAv(r):
    """
    r: array
    returns: average density in cars/km
    """
    return dx/L*r.sum()

def Vav(r):
    """
    r : array
    returns: average velocity in m/s
    """
    return v_max*(1-rhoAv(r)/rho_max)/3.6
```
